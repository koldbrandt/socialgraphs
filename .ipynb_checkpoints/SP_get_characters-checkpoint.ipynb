### Get All Characters
First we get all characters from the characters site on the fandom wiki

```python
baseurl = "https://southpark.fandom.com/api.php?"
action = "action=query"
title = "titles=Portal:Characters"
content = "prop=revisions&rvprop=content"
dataformat ="format=json"
someOtherThing = "rvslots=*"
query = "{}{}&{}&{}&{}&{}".format(baseurl, action, content, title, dataformat,someOtherThing)

# Making a get request
response = requests.get(query)
```

Getting the text from the json file that gets returned from the request

```python
characters_pages_all = response.json()['query']['pages']['3590']['revisions'][0]['slots']['main']['*']
```

Using regex to get all character links on the site and cleaning to only get the name

```python
characters_all  = re.findall('\[\[[a-zA-Z0-9\'\\ -.]*\]\]', characters_pages_all)
characters_all = list(set([w.replace('[[', '').replace(']]', '') for w in characters_all]))
```

```python
len(characters_all)
```

This gives 1837 characters, there are however some flaws. In the top of the page a link to "South Park" is handled as a character. We have to remove that manually. There are also groups of characters that are listed in the character page. First we can remove "South Park". We will come back to the groups later, since they will get the category group and they will therefore be easier to spot later on. 

```python
characters_all = [ch for ch in characters_all if ch != "South Park"]
print("South Park" in characters_all)
```

## Download character pages
From the list of characters it is now possible to download all character pages.  
Startinhg by creating a dataframe with the characters.

```python
df_characters = pd.DataFrame(characters_all, columns=['Name'])
```

```python
df_characters.head()
```

The code for downloading each character page is in the two functions below. It handles if the response get a redirect from a charcter that has another name than listed in the list of characters. There is also a special case with jerome, because he has a dark version of himself. This is not handled by the redirect, so this is done manually. 

```python
def downloadCharacterPage(character):
    character = character.replace("&", "%26")
    jerome = False
    if character == "Darth Chef":
        character = "Jerome \"Chef\" McElroy#Darth Chef"
        jerome = True
    baseurl = "https://southpark.fandom.com/api.php?"
    action = "action=query"
    title = "titles=" + character.replace(" ", "_")
    content = "prop=revisions&rvprop=content"
    dataformat ="format=json"
    someOtherThing = "rvslots=*"
    query = "{}{}&{}&{}&{}&{}".format(baseurl, action, content, title, dataformat,someOtherThing)
    response = requests.get(query)
    if jerome:
         return response.text
    if re.search('\#REDIRECT', response.text) == None:
        return response.json()
    newTitle = re.findall(r'[[*[A-Za-z0-9 ., _()]*\]]', response.text)
    newTitle = newTitle[0].split('[[')
    print("old char :", character, " | new char: ", newTitle[1][:-2])
    return downloadCharacterPage(newTitle[1][:-2])
    

def downloadCharacter(char):
    page = downloadCharacterPage(char)
    if not isinstance(page, str):
        key = list(page['query']['pages'].keys())
        page = page['query']['pages'][key[0]]['revisions'][0]['slots']['main']['*'] 
    with open('southpark_pages/' + char.replace("\"", "") + '.txt', 'w', encoding='utf-8') as f:
        f.write(page)
```

## Categories

At the bottom of each character page its categories are listed. These categories will now be pulled and stored in the character dataframe.

```python
categories_list=[]
for index, Row in df_characters.T.iteritems():
    name = Row.Name
    with open('southpark_char_pages/' + name.replace("\"", "") + ".txt", 'r', encoding='utf-8' ) as file:
        data = file.read()
        categories_list.append(re.findall(r'(?<=\[\[Category:)(.*?)(?=\])',data))
df_characters['Categories'] = categories_list
```

This list below is taken from the website, which contains all the categories used there. This list had to be generated manually, as each page does not include specifically what category a character is in. For instance, all the 4th graders does not have "school characters" as a category on their page, even though they are clearly school characters. This is not the case with celebrities, where every page of a celebrity, also had a corresponding celebrity category. To sum up, the inconsisency of the wiki page made it necessary to explicitly write the categories ourselves.  

```python
wikiCategories=['School Characters','Kindergartners','1st-3rd Graders','4th Graders','5th-9th Graders','School Faculty','Creatures','Aliens', 'Animals', 'Monsters',
 'Other', 'Alter Egos ','Crossover characters','Internet Trolls',"Paris Hilton's Pets ",'Raisins Employees','Animated South Park Fans ','Unnamed Characters', 
  'Groups','Business Celebrities','Celebrities','Criminal Celebrities','Fictional Celebrities','Historical Celebrities','Internet Celebrities',
  'Music Celebrities','Political Celebrities','Religious Figures','Sports Celebrities','Television Celebrities']
```

The intersection betweenthe df_characters and the wiki categories list is found, and added as a common value in the df. If no category is found, the character gets the category: uncategorized. As one character could have more than one category, the categories are split into subcategories. 

```python
def compare(categoriesx, categoryList):
    categoryList=[x.lower() for x in categoryList]
    categoriesx=[x.lower() for x in categoriesx]
    return list(set(categoryList).intersection(categoriesx))

df_characters['commonValues'] = df_characters.Categories.apply(lambda x: compare(x,wikiCategories))
df_characters['commonValues'] = df_characters['commonValues'].apply(lambda y: ["uncategorized"] if len(y)==0 else y)
df_characters=df_characters.rename(columns={"commonValues": "mainCategories"})
#Skal potentielt fjernes
df_characters = df_characters.replace({'\"': ''}, regex=True)
dftemp = df_characters["mainCategories"].apply(pd.Series)
df_characters = pd.merge(df_characters, dftemp, left_index=True, right_index=True)
df_characters = df_characters.rename(columns={0: "subCategory1", 1: "subCategory2",2: "subCategory3", 3: "subCategory4"})
df_characters = df_characters.drop(columns="Categories")
```

All characters with the category group can now be removed. This is done because multiple characters are part of a group and are not a single character.

```python
df_characters = df_characters.drop(df_characters[df_characters.subCategory1 == 'groups'].index)
```

## Get links 
The character pages has links to other characters. These links will be stored in the dataframe and from these links a digraph can be created.   
The function to find the list of characters that get linked to is created.

```python
def get_characters(char):
    with open('southpark_char_pages/' + char.replace("\"", "") + ".txt", 'r', encoding='utf-8' ) as file:
        data = file.read()
    potential_links = re.findall("\[\[*[A-Za-z _()\|\']*\]\]",data)
    potential_links = [x.split('[') for x in potential_links]
    potential_links = [x[2].split(']') for x in potential_links]
    potential_links = [x[0] for x in potential_links]
    potential_links = [x.split('|')[0] for x in potential_links]
    # removing doubles 
    characters = []
    potential_links = list(set(potential_links))
    # check to see if it is a character or another random link
    for pot in potential_links:
        if pot in df_characters.Name.values:
            characters.append(pot)
    return characters
```

Using an apply on the characters dataframe to create a new column with the list od characters that the character links to.

```python
df_characters['character_links'] = df_characters.Name.apply(lambda x : get_characters(x))
```

## Creating a network
It is now possible to create a network from this. 

```python
def make_connection(characterlist, name, G):
    for i in characterlist:
        G.add_edge(name, i)
    return G
```

Creating the network

```python
G = nx.DiGraph()
for index, row in df_characters.iterrows():
    make_connection(row['character_links'], row['Name'], G)
```

Some basic stats for the network

```python
def printBasicStats(G):
    print("Basic stats for the gcc network:")
    N = len(G.nodes())
    K = sum(dict(G.degree()).values())/N
    d = log(N)/log(K)
    print("Number of nodes: ", N)
    print("Number of edges: ", len(G.edges()))
    print("Average degree: ", K)
    print("Average distance between 2 nodes: ", d)
    print("\n5 most in degree: " )
    in_deg = sorted(G.in_degree, key=lambda x: x[1], reverse=True)
    for node, degree in in_deg[0:5]:
        print(f'{node}: {degree} edges')
    print("\n5 most out degree: " )
    out_deg = sorted(G.out_degree, key=lambda x: x[1], reverse=True)
    for node, degree in out_deg[0:5]:
        print(f'{node}: {degree} edges')
    
printBasicStats(G)
```

Saving the directed graph

```python
nx.write_gpickle(G, 'SouthParkDirected.gpickle')
```

Getting the gcc and saving it borh for the directed and undirected

```python
G_large = G.subgraph(max(nx.weakly_connected_components(G), key=len)).copy()
nx.write_gpickle(G_large, 'SouthParkDirectedGCC.gpickle')
gcc = G_large.to_undirected()
nx.write_gpickle(gcc, 'SouthParkUnirectedGCC.gpickle')
```

## Plotting GCC
Using force atlas to plot. Setting the nodesize from how many degrees the node has, and the color from its category.

```python
# set node size depending on degree
d = dict(gcc.degree)
node_sizes = [8*v for v in d.values()]
```

Creating a dictionary with a color for each category.

```python
num_colors=len(set(wikiCategories))
colors=list(mcolors.CSS4_COLORS.keys())
cluster_colors=random.choices(colors, k=num_colors)

categories = list(set(df_characters.subCategory1.values))

color_dict = {}
for i in range(len(categories)):
    color_dict[categories[i]] = cluster_colors[i]
```

Creating a list with the color for each node

```python
node_color = []
for node in gcc.nodes():
    cat = df_characters[df_characters.Name == node].subCategory1.values[0]
    node_color.append(color_dict[cat])
```

```python
# Setup FA2 w. parameters
forceatlas2 = ForceAtlas2(
    # Behavior alternatives
    outboundAttractionDistribution=True,    # Dissuade hubs
    edgeWeightInfluence=1.0,

    # Performance
    jitterTolerance=1.0,  # Tolerance
    barnesHutOptimize=True,
    barnesHutTheta=1.2,

    # Tuning
    scalingRatio=1.0,
    strongGravityMode=False,
    gravity=0.2,

    # Log
    verbose=True

    
)

# compute node positions with the fa algorithm
node_pos = forceatlas2.forceatlas2_networkx_layout(gcc, pos=None, iterations=2000)
```

```python
plt.figure(figsize=(15, 10))
nx.draw_networkx_nodes(gcc, node_pos, nodelist=d.keys(), node_size=node_sizes, alpha=0.4, node_color = node_color)
nx.draw_networkx_edges(gcc, node_pos, alpha=0.1)
```

## Clean character pages

To do the wordclouds we need the pages cleaned up. We use the following method to do so. 

```python
def save_cleaned_text(character):
    with open('southpark_char_pages/' + character.replace("\"", "") + ".txt", 'r', encoding='utf-8' ) as file:
        data = file.read()
    test = re.sub('{{[a-zA-Z0-9\'\\ /\"-.|=:<>? \r?\n \[\]]*}}', '', data)
    test = re.sub("==[a-zA-Z0-9\'\\ -.:<>?]*==",'',test)
    test = re.sub("===[a-zA-Z0-9\'\\ -.:<>?]*===",'',test)
    test = re.sub("\[\[[a-zA-Z0-9\'\\ -.:()|<>? \u4e00-\u9fff]*\]\]",'',test)
    test = re.sub("\\n",' ',test)
    test = re.sub(r'http\S+', '', test)
    test = re.sub("<[^>]*>",'',test)
    test = re.sub("[a-zA-Z 0-9-.]*.png",'',test)
    test = re.sub("[a-zA-Z 0-9-.]*.jpg",'',test)
    test = re.sub("seen|episode|appears|South|park",'',test)
    test = re.sub('[^A-Za-z ]+', '', test)
    with open('southpark_pages_clean/' + character.replace("\"", "") + '.txt', 'w', encoding='utf-8') as f:
        f.write(test)
```
